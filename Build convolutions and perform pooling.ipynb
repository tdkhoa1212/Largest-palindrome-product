```python
import cv2
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt

i = misc.ascent() #Get an 8-bit grayscale bit-depth, 512 x 512 derived image for easy use in demos

plt.grid('false')
plt.gray()
plt.axis('off')
plt.imshow(i)
plt.show()
```

```python
i_tranform = np.copy(i)

size_x = i_tranform.shape[0]
size_y = i_tranform.shape[1]

# 1: make a convolution matrix (or kernel) as a 3x3 array:
'''Using [-1,-2,-1,0,0,0,1,2,1] gives you horizontal lines:==================================='''
filter = [ [-1, -2, -1], [0, 0, 0], [1, 2, 1] ]
weight = 1

# 2: load the new value into the transformed image:
for x in range(1, size_x-1):
    for y in range(1, size_y-1):
        output_pixel = 0.0
        
        output_pixel = output_pixel + ( i[x-1][y-1] * filter[0][0] )
        output_pixel = output_pixel + ( i[x][y-1] * filter[0][1])
        output_pixel = output_pixel + ( i[x+1][y-1] * filter[0][2])
        
        output_pixel = output_pixel + ( i[x-1][y] * filter[1][0])
        output_pixel = output_pixel + ( i[x][y] * filter[1][1])
        output_pixel = output_pixel + ( i[x+1][y] * filter[1][2])
        
        output_pixel = output_pixel + ( i[x-1][y+1] * filter[2][0])
        output_pixel = output_pixel + ( i[x][y+1] * filter[2][1])
        output_pixel = output_pixel + ( i[x+1][y+1] * filter[2][2])
        output_pixel = output_pixel * weight
        
        if output_pixel < 0: output_pixel = 0
        if output_pixel > 255: output_pixel = 255
            
        i_tranform[x][y] = output_pixel

plt.imshow(i_tranform)
plt.show() 
```

```python
new_size_x = int(size_x/2)
new_size_y = int(size_y/2)
new_image = np.zeros((new_size_x, new_size_y))

for x in range(0, size_x, 2):
    for y in range(0, size_y, 2):
        pixels = []
        pixels.append(i_tranform[x][y])
        pixels.append(i_tranform[x+1][y])
        pixels.append(i_tranform[x][y+1])
        pixels.append(i_tranform[x+1][y+1])
        pixels.sort(reverse=True)
        new_image[int(x/2)][int(y/2)] = pixels[0]
        
# Plot the image. Note the size of the axes -- now 256 pixels instead of 512
plt.imshow(new_image)
#plt.axis('off')
plt.show()
```
